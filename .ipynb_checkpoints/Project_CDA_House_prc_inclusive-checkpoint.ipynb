## CDA Final Project : House Price prediction

```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.model_selection import RandomizedSearchCV
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.legend_handler import HandlerLine2D
from sklearn.metrics import r2_score
import xgboost
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import train_test_split
import operator
from sklearn.ensemble import RandomForestRegressor
from copy import deepcopy
```

### Loading the Ames (Iowa) house pricing dataset

```python
"""Inputting the training and the test dataset"""
# Point to note - No need for data normalization as such for Random forrest regressor
Training_set = pd.read_csv('./train.csv', index_col = 'Id')
Testing_set = pd.read_csv('./test.csv', index_col = 'Id')

#IDs of testing set
Testing_set_IDs = list(Testing_set.index)

Y_train = Training_set['SalePrice']
del Training_set['SalePrice']
```

```python
# Combining the dataset
Data_set = pd.concat([Training_set,Testing_set])
```

```python
Data_set.head()
```

### Cleaning up the dataset removing columns with excessive NAs 

```python
"""Analysing number of NAs present in the combined dataset"""
cols_with_na_vals = [cols for cols in Data_set.columns if Data_set[cols].isnull().sum()>=1]
print('\nTotal cols with NA vals', len(cols_with_na_vals), '\n')

na_vals_cols_dict = {cols_with_na_vals[i]:np.round(Data_set[cols_with_na_vals[i]].isnull().mean()*100,2) for i in range(len(cols_with_na_vals))}
na_vals_cols_dict = dict(sorted(na_vals_cols_dict.items(), key=operator.itemgetter(1),reverse=True))

for k,v in na_vals_cols_dict.items():
    print(k,' - ', v, '% of missing value')
```

```python
# Printing columns with more than 80% NA vals
mt_80na_cols = [cols for cols in cols_with_na_vals if Data_set[cols].isnull().mean()*100 >= 80]
print(mt_80na_cols)
```

```python
"""Dropping 'PoolQC', 'Fence', 'MiscFeature' beacuse according to feature importance(performed in the end), they are not really that important
and also, they have a bunch of NA vals""" 
Data_set.drop(['PoolQC', 'Fence', 'MiscFeature'], axis=1, inplace=True)
```

### Reordering Columns (separating out Alphbetical and Numerical columns)

```python
"""Reordering columns"""
Coltype_dict = pd.DataFrame(Data_set.dtypes)
Coltype_dict.reset_index(inplace = True)
Coltype_dict = Coltype_dict.rename(columns={0:"Datatype", "index":"Colname"})
Coltype_dict["Datatype"]= Coltype_dict["Datatype"].astype(str) 
Coltype_dict = dict(zip(Coltype_dict['Colname'],Coltype_dict['Datatype']))
```

```python
Numerical_cols = [k for k,v in Coltype_dict.items() if (v == 'int64' or v == 'float64')]
NonNumer_cols = [k for k,v in Coltype_dict.items() if v == 'object']
Rearngd_cols = Numerical_cols + NonNumer_cols

Data_set = Data_set.reindex(columns=Rearngd_cols)
```

```python
"""Replacing nans with zeros"""
Data_set = Data_set.fillna(0)
```

```python
Data_set.columns
```

### Encoding the alphabetical columns using sklearn label encoder

```python
"""Using Label encoder to transform string cols"""
LE = preprocessing.LabelEncoder()

# Encoding the dataset
Strcols = Data_set.loc[:, 'MSZoning':]
Strcols = Strcols.astype(str)
Strcols = Strcols.apply(LE.fit_transform)
Data_set = pd.concat([Data_set.loc[:, :'YrSold'], Strcols], axis=1)
```

### Splitting the training and testing datasets

```python
train_lngth = Training_set.shape[0]
```

```python
"""Train test split again"""
Training_set = deepcopy(Data_set.iloc[:train_lngth, :])
Testing_set = deepcopy(Data_set.iloc[train_lngth:, :])
```

### Implementing XGBoost with n fold crossvalidation

```python
"""Implementing XGBoost"""

skf = StratifiedKFold(n_splits = 5, shuffle = True)
random_grid = dict(learning_rate = [0.05, 0.1, 0.15, 0.2, 0.25, 0.3], max_depth = [3,4,5,6,8,10,12,15],
                 min_child_weight = [1,3,5,7], gamma = [0.0,0.1,0.2,0.3,0.4], colsample_bytree = [0.3,0.4,0.5,0.6])
xgboost_tuned = RandomizedSearchCV(estimator = xgboost.XGBRegressor(), param_distributions = random_grid, n_iter = 200, cv=10, scoring = 'r2', verbose=2, n_jobs = -1)
xgboost_tuned.fit(Training_set, Y_train)   #skf.split(Training_set, Y_train)
print(xgboost_tuned.best_estimator_)
```

```python
from xgboost import plot_tree
fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize =(20, 40))
plot_tree(xgboost_tuned.best_estimator_, filled=True, ax = ax)
ax.set_title('Visualising the fitted XGBoost tree', size = 25)
plt.show()
```

### In sample R-squarred

```python
# Training set predicted values
Y_train_pred = xgboost_tuned.predict(Training_set)

# Training set R-squarred (In sample R squarred)
Insample_r2 = r2_score(Y_train, Y_train_pred)
print('In sample R-squarred -', Insample_r2)
```

### Implementing Random forest

```python
"""Implementing Random forest"""
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]
max_features = ['auto', 'sqrt']
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
max_depth.append(None)
min_samples_split = [2, 5, 10]
min_samples_leaf = [1, 2, 4]
bootstrap = [True, False]

random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}

RandomForest_Tuned = RandomizedSearchCV(estimator = RandomForestRegressor(), param_distributions = random_grid, n_iter = 100, cv = 5, verbose=2, random_state=42, n_jobs = -1)
RandomForest_Tuned.fit(Training_set, Y_train)
print(RandomForest_Tuned.best_estimator_)
```

### In sample R-squarred

```python
# Training set predicted values
Y_train_pred_RF = RandomForest_Tuned.predict(Training_set)

# Training set R-squarred (In sample R squarred)
Insample_r2_RF = r2_score(Y_train, Y_train_pred_RF)
print('In sample R-squarred -', Insample_r2_RF)
```

### Regression metrics 

```python
# Testing set predicted values
Y_test_pred_RF = RandomForest_Tuned.predict(Testing_set)
```

```python
# Explained variance score
explained_variance_score(Y_test, Y_test_pred_RF)
```

```python
# Feature Importance for Random Forrest
model = RandomForest_Tuned.best_estimator_
importance = model.feature_importances_

importance_dict = {Testing_set.columns[i]:importance[i] for i in range(len(importance))}

import operator
importance_dict = dict(sorted(importance_dict.items(), key=operator.itemgetter(1),reverse=True))

# plotting feature importance
plt.figure(figsize=(15, 15))
plt.barh(list(importance_dict.keys()), list(importance_dict.values()))
plt.xlabel('Performance', size = 15)
plt.ylabel('Features', size = 15)
plt.title('Analysing feature importance', size = 25)
plt.show()
```

### Comments

It can be noted that XGBoost is performing way better as compared to random forrest algorithm. To perform 10 fold cross validation, random forrest algorithm takes about 90 minutes to complete whereas XGBoost algorithm takes merely about 10 minutes. 

Furthermore, root mean squarred log error is much lower for XGBoost as compared to Random Forrest algorithm. Based on time taken to train the algorithms, it can be said that it is much easier to further tune the parameters for XGBoost as compared to Random forrest algorithm.

As far as feature importance is concerned, for Random forrest algorithm, variables - 'OverallQual', 'GrLivArea', 'GarageCars' and 'YearBuilt' seem to be the most important features while constructing the model. For XGBoost algorithm, variables - 'OverallQual', 'KitchenQual', 'GarageCars' and 'ExterQual' are the most important features for regression. This tells us that 'OverallQual' is the most important feature for correctly predicting SalePrice. Earlier during exploratory data analysis, we also observed that correlation for this feature with that of SalePrice is the highest as well.
